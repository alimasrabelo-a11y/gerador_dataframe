```python
#-------------------------Gerador csv-------------------------------------

#importa do biblioteca
import pandas as pd #importando pandas
import numpy as np # importando numpy

#declarando constantes
Tipo_atividade=["Compra","Visualização","Carrinho"]

#criando catálogo de 50 produtos
qtd_produto=50
lista_produto=np.random.randint(100,999 , size=qtd_produto)

#criando dataframe vazio com colunas desejadas
df = df[["ID_transacao", "ID_cliente", "ID_produto", "Valor", "Data_Hora", "Atividade"]]

#funcao gerar números aleatórios e preencher os dados do csv
#criando sincronização para o produto sempre tenha o mesmo valor
catalogo={}
for prod in lista_produto:
  v_inteiro=np.random.randint(10,500) #gerando valor em reais
  v_centavo=np.random.randint(0,99)#gerando valor centavos
  catalogo[prod]=f"R${v_inteiro},{v_centavo:02d}"#edita o texto para reais


def aleatorio (qtd=10):

#laco de repetição limitado
  for n in range (qtd):
    Id_transacao=np.random.randint(100, 999)#gerando numero aleatorio de 3 digitos
    Id_cliente=np.random.randint(100, 999)#gerando numero aleatorio de 3 digitos
    hora=f"{np.random.randint(0,23):02d}:{np.random.randint(0,59):02d}"# hora aleatoria
    #sorteia um produto que ja estaja no catalogo
    Id_produto=np.random.choice(lista_produto)

    #busca o valor fixo do produto naquele catalogo
    valor_fixo=catalogo[Id_produto]


    #gerar data aleatoria
    dia=np.random.randint(1,29)
    mes=np.random.randint(1,13)
    ano=np.random.randint(2013,2026)
    data=pd.to_datetime(f"{dia:02d}/{mes:02d}/{ano}", dayfirst=True)
    atividade=np.random.choice(Tipo_atividade)#comando seleção aleatoria

    #formatar para o padrão brasileiro
    data_str = data.strftime("%d/%m/%Y")
    #juntar data e hora
    data_hora=f"{data_str} {hora}"

    #fazendo valor aparecer apenas em compras



    #preenchendo informações do dataframe
    df.loc[len(df)]=[ Id_transacao, Id_cliente,Id_produto,valor_fixo, data_hora , atividade]

#gerar registros
aleatorio(10000)#escolha a quantidade de registros

#criando o CSV
df.to_csv("dados_transacionais.csv", index=False, encoding="utf-8")

print(df)

#salvar direto no computador
from google.colab import files
files.download("dados_transacionais.csv")
```
